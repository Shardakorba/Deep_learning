```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,50,100)
```

```python
x.shape
```

```python
y  = 2*x+4
```

```python
y.shape
```

```python
plt.scatter(x, y,s=2)
plt.xlabel("X axis")
plt.ylabel("y axis")
```

```python
np.random.seed(40)
N = len(y)
L = 100
factor = 8
noise = np.random.randint(-factor*L,factor*L,N)/L
```

```python
y0 = noise + y
```

```python
plt.plot(x,y,c="blue")
plt.scatter(x,y0,s=2,c="green")
```

```python
b = np.sum((x-np.mean(x))*(y0-np.mean(y0)))/np.sum((x-np.mean(x))*(x-np.mean(x)))
```

```python
a = np.mean(y0)-b*np.mean(x)
```

```python
y1 = a+b*x
```

```python
plt.plot(x,y,c="blue")
plt.scatter(x,y0,s=2,c="green")
plt.plot(x,y1,c= "red")
```
